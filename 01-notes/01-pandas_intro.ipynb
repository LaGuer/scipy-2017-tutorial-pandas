# grouped and aggregated calculations

Look at all the cool things dataframes and series objects can do!

- https://pandas.pydata.org/pandas-docs/stable/generated/pandas.DataFrame.html
- https://pandas.pydata.org/pandas-docs/stable/generated/pandas.Series.html

```python
df.head(n=10)
```

```python
df.groupby('Year')
```

```python
df.groupby('Year')['Product Name']
```

```python
df.groupby('Year')['Value'].mean()
```

```python
df.groupby('Year')[['Product Name', 'Value']].mean()
```

```python
df.groupby(['Year', 'Country/Region Name'])[['Product Name', 'Value']].mean()
```

```python
df.groupby('Product Name')['Country/Region Name'].nunique()
```

```python
import matplotlib.pyplot
%matplotlib inline
```

```python
gyle = df.groupby('Year')['Value'].mean()
gyle
```

```python
gyle.plot()
```
