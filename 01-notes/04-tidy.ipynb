##  Multiple variables are stored in one column.

````python
# tb = pd.read_csv('../data/tb.csv')
tb = pd.read_csv('https://raw.githubusercontent.com/LaGuer/JupyterDataScienceWorkflow/master/dataset/oecd-CRSW-crop-production.csv')
tb.head()
````

````python
# ebola = pd.read_csv('../data/ebola_country_timeseries.csv')
crop2 = pd.read_csv('https://raw.githubusercontent.com/LaGuer/JupyterDataScienceWorkflow/master/dataset/oecd-CRSW-crop-production.csv')
````

````python
crop2.head()
````

````python
# first let's melt the data down
crop_long = crop2.melt(id_vars=['LOCATION', 'INDICATOR', 'SUBJECT', 'MEASURE', 'FREQUENCY', 'TIME', 'Flag Codes'],
                        value_name='VolVal',
                        var_name='Volume')
crop_long.head()
````

````python
var_split = crop_long['SUBJECT'].str.split('_')
var_split
````

````python
type(var_split)
````

````python
var_split[0]
````

````python
var_split[0][0]
````

````python
var_split[0][1]
````

````python
# save each part to a separate variable
status_values = var_split.str.get(0)
country_values = var_split.str.get(1)
````

````python
status_values.head()
````

````python
country_values.head()
````

````python
# assign the parts to new dataframe columns
crop_long['status'] = status_values
crop_long['country'] = country_values
````

````python
crop_long.head()
````

### above in a single step

````python
variable_split = crop_long['SUBJECT'].str.split('_', expand=True)
variable_split.head()
````

````python
# variable_split.columns = ['SUBJECT', 'LOCATION']
variable_split.columns = ['SUBJECT']
````

````python
rice = pd.concat([crop_long, variable_split], axis=1)
rice.head()
````

## Variables in both rows and columns

````python
weather = pd.read_csv('../data/weather.csv')
````

````python
weather.head()
````

````python
weather_melt = pd.melt(weather,
                      id_vars=['id', 'year', 'month', 'element'],
                      var_name='day',
                      value_name='temp')
````

````python
weather_melt.head()
````

````python
weather_tidy = weather_melt.pivot_table(
    index=['id', 'year', 'month', 'day'],
    columns='element',
    values='temp')
weather_tidy.head()
````

````python
weather_flat = weather_tidy.reset_index()
weather_flat.head()
````
