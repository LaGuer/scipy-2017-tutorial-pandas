```python
## use agg to call functions from other libraries
## or even functions you write yourself
import numpy as np

# these 2 do the same thing
rice.groupby('SUBJECT')['Value'].aggregate(np.mean)
wheat.groupby('Product Name')['Value'].agg(np.mean)
```

```python
def my_mean(values):
    n = len(values)
    s = np.sum(values)
    return s / n
```

```python
rice.groupby('SUBJECT')['Value'].agg(my_mean)
```

```python
# multiple functions
rice.groupby('TIME')['Value'].agg([
    np.count_nonzero,
    np.mean,
    np.std
])
```

```python
rice.groupby('TIME')['Value'].agg({
    'ncount': np.count_nonzero,
    'mean': np.mean,
    'std': np.std
}).reset_index()
```

http://pandas.pydata.org/pandas-docs/version/0.20/whatsnew.html#deprecate-groupby-agg-with-a-dictionary-when-renaming


```python
wheat.groupby('Year')['Value'].\
    agg([np.count_nonzero, np.mean, np.std]).\
    rename(columns={'count_nonzero': 'count',
                   'mean': 'avg',
                   'std': 'std_dev'}).\
    reset_index()
```
