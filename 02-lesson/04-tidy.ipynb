Population Density per square km 
-------------------------------------------------

```python
popdensity = pd.read_csv('https://raw.githubusercontent.com/LaGuer/JupyterDataScienceWorkflow/master/dataset/population_density_per_square_km.csv')
```

```python
popdensity.head()
```

```python
popdensity_melt = pd.melt(popdensity,
                     id_vars=['1950', '2000'],
                     var_name='cd_country',
                     value_name='country')
```

```python
popdensity_melt
```

```python
var_split = popdensity_melt['cd_country'].str.split('_')
```

```python
status_values = var_split.str.get(0)
country_values = var_split.str.get(1)
```

```python
popdensity_melt['status'] = status_values
popdensity_melt['country'] = country_values
```

```python
popdensity_melt.head()
```

```python
variable_split = popdensity_melt['cd_country'].str.split('_', expand=True)
variable_split
```

```python
variable_split.columns = ['status1', 'country1']
variable_split.head()
```

```python
popdensity_clean = pd.concat([popdensity_melt, variable_split], axis=1)
```

```python
popdensity_clean.head()
```

```python
ictineu = pd.read_csv('https://raw.githubusercontent.com/LaGuer/JupyterDataScienceWorkflow/master/dataset/eu-ict-expenditures.csv')
ictineu.head()
```

```python
ictineu.columns
```

```python
ictineu_melt = pd.melt(ictineu,
                      id_vars=['id', 'year', 'month', 'element'],
                      var_name='day',
                      value_name='temp')
```

```python
ictineu_melt.head()
```

```python
ictineu_tidy = ictineu_melt.pivot_table(
    index=['id', 'year', 'month', 'day'],
    columns='element',
    values='temp'
).reset_index()
ictineu_tidy
```
